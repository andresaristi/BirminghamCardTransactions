## 1. Business Questions

For this project we are using a historical dataset wich includes a collection of purchase card transactions for the Birmingham City Council.

There are several important tasks which can by performed in order to get the most of the data presented.

### 1.1 We list the possible tasks that could be performed and are interesting enough to be analysed and taked into account for this project:

1. (Clustering) Discovering profiles.
2. (Anomalies detection) Unusual activities.
3. (Forecasting) Predicting future transactional behaviors (Next purchase, Expenditures forecasting, etc.)
4. (Creativity) Stating a problem that may be solved using the available data.

### 1.2 By taking into account these tasks we establish a series of important questions which could be answered by following the Data Analysis Process stated previously.

1. Clustering: a) which are the all around purchase card transactions' profiles?, b) which are the credit card customers' profiles?, c) which are the merchants' profiles?
2. Anomalies detection: a) which transactions are unusual and may be fraudulent?
3. Forecasting: what would the next purchase be for a particular client?, b) how much would a client expend in the next period (days, month, years)?, c) how much would be the income for a particular merchant in the next period (days, months, years)?
4. Creativity: Classification: Taking into account a time frame: a) what would be, for a card transaction, the most probable type of consumption (Vehicle Fuel, Hospitality, Food, Accomodation, etc.), b) what would be, for a customer, the most probable type of consumption (Vehicle Fuel, Hospitality, Food, Accomodation, etc.), c) what would be, for a card transaction, the most probable company in which he/she would expend his/her money?, d) what would be, for a customer, the most probable company in which he/she would expend his/her money?, e) what would be the most probable card transaction associated directorate?, f) what would be the most probable card transaction Value Added Tax (VAT)?,

### 2.2 Merging the data by years

We define a function that will take the months, the years, all the xls files and the path to where the files are saved in order to obtain a list of the dataframes that will include the information per year.

```python
def merging_xls_in_df(months, years, files, path):
    dfs = list()
    for y in years:
        df = pd.DataFrame()
        for m in months:
            for f in files:
                if f.endswith(".xls") and m in f and y in f:
                    df = df.append(pd.read_excel(path+'/'+f, parse_dates = True))
        dfs.append(df)
    return dfs
```

We declare the lists of months, years and files as well as the path in which the files will be found in order to be open and included in each particular dataframe per year

```python
months = ['jan','feb','mar','apr','may','jun','jul','aug','sept','oct','nov','dec']
years = ['2014','2015','2016','2017','2018']
destdir = '/Users/andresaristi/Documents/BirminghamCardTransactions/dataset'
files = [ f for f in os.listdir(destdir) if os.path.isfile(os.path.join(destdir,f)) ] 
```

We create a function to do more feature engineering by creating columns related to the main description of card transactions per client.

```python
def add_trans_description(descriptions, data, variable, name):
    df = pd.DataFrame()
    i=0
    for descr in descriptions:
        df1 = pd.DataFrame()
        print(descr)
        #exec("%s = %s" % (des, descr)) (
        df1 = data.groupby(variable).agg(descr=(name, lambda val: (val == descr).sum()))
        df = pd.concat([df,df1], axis=1)
    return df
```

#### Interpretation of k-means using k=3

Even though we used a k=3, there is one cluster with only one data point, so when analyzing the kernel density plots we are unable to see cluster number 1.

Nevertheless, we, by using these density plots, are going to understand the different characteristics of each cluster according to the observations which belong to them.

By observing the plots we are able to understand the type and amount of card transaction purchases of each group of clients.

<b>Cluster 0:</b> the most densely populated. The clients belonging to this segment make transactions with small accumulative gross quantities. In general do not spend much in office equipments and just some of their card transactions are related to services.

<b>Cluster 2:</b> the accumulative gross amount of the clients belonging to this group is considerably higher than the first cluster, although the population is much smaller. They make purchases related to office equipments.

```python
var_num = data2014_kmeans.columns[0:3]
fig = plt.figure(figsize=(18,15))
i=1
for var in var_num:
    ax = fig.add_subplot(math.ceil(len(var_num)/2), 2, i)
    sns.kdeplot(data2014_kmeans.loc[data2014_kmeans.Cluster==0][var], shade=True, color='r', ax=ax);
    sns.kdeplot(data2014_kmeans.loc[data2014_kmeans.Cluster==1][var], shade=True, color='g', ax=ax);
    sns.kdeplot(data2014_kmeans.loc[data2014_kmeans.Cluster==2][var], shade=True, color='b', ax=ax);
    plt.title(var)
    plt.legend(['Cluster 0', 'Cluster 2', 'Cluster 1'])
    i+=1
```

We will see the scatterplots to better understand the differences

Using the scatterplot we found out that the only data point in cluster 1 seems like an atypical client which spends too much.

#### Determining the right k

#### Elbow method

By using the Within Sum of Squares (WSS) values we will create a plot by which we will apply the elbow method, and  cluster using different k values.

```python
WSSs = []
for i in range(1,15) :
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df2014_kmeans)
    WSSs.append(km.inertia_)
WSSs
```

We found out that with this technique the ideal k=2

```python
plt.plot(range(1, 15), WSSs)
```

#### Silhouette method

By using the silhoutte method we will try to find a different approach to the number of clusters that we will need for our dataset. We will be using k=2, k=3 and k=4, k=5, k=6

```python
k_values = [2,3,4,5,6,7,8,9]
for k in k_values:
    
    plt.figure(figsize=(8,8))

    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(df2014_kmeans)
    y_clusters = kmeans.labels_
    cluster_labels = np.unique(y_clusters)

    silhouette_points= silhouette_samples(df2014_kmeans, y_clusters, metric='euclidean')

    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    colors = ['r', 'g', 'b', 'y', '0', 'c', 'm', 'k','aquamarine']
    for i, c in enumerate(cluster_labels):
        silhouette_points_c = silhouette_points[y_clusters == c]
        silhouette_points_c.sort()
        y_ax_upper += len(silhouette_points_c)
        color = colors[i]
        plt.barh(range(y_ax_lower, y_ax_upper), silhouette_points_c, height=1.0, 
                 edgecolor='none', color=color)

        yticks.append((y_ax_lower + y_ax_upper) / 2.)
        y_ax_lower += len(silhouette_points_c)
    
    silhouette_mean = np.mean(silhouette_points)
    plt.axvline(silhouette_mean, color="black", linestyle="--") 

    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette Coefficient')

    plt.tight_layout()
    plt.show()
```

After analyzing the results from the silhouette method we found out that even though the clusters for k=2 have considerable different widths, one surpasses the Silhouette Coefficient mean and the other is close by.

Meanwhile, the other possibilities include clusters with negative values for their Silhouette coefficients.

#### Calinski-Harabasz

Finally, in order to be thorough we will apply the The Calinski-Harabasz criterion to have a new tool in order to choose the right k.

```python
CHs = []
for i in range(2,15) :
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df2014_kmeans)
    CH = calinski_harabasz_score(df2014_kmeans, km.labels_) 
    CHs.append(CH)
CHs
```

```python
plt.plot(range(2, 15), CHs)
```

Accoring to de Calinski-Harabasz criterion we found out that the ideal k would be 8, but it does not agree with the two previous heuristics. Taking into account that we are profiling clients, 2 segments may not be an ideal number of groups to establish marketing campaigns for those groups. 

We will use k = 2 and compare it again with k = 8 (as the Variance Ratio Criterion suggests)

#### Profiling clients with k-means using k=8

Although we already profiled the clients using k-means with k=2 and we found out by analyzing the elbow and the silhouette methods that 8 is not an ideal number for the amount of clusters to be used, it is important to take into account the Calinski-Harabasz criterion and the business intelligence.

Even if 2 clusters seem ideal for the obtained data, from the business point it would be advisable to identify more clusters in order, for example, to profile clients and be able to establish specific marketing campaigns for each of these clients in order for them to increase their purchases (not only related to the amount of the transactions, but to the quantity paid for a product). It is important to find out different groups so the company or companies may offer special deals for products that they may want.

Therefore, by observing the plots we are going to be able to understand the type and amount of card transaction purchases of each group of clients.

<b>Cluster 0:</b> not very densely populated. Involves those clients that not spend much money in purchase card transactions. The few transactions that they make involve mostly office equipments. Their transactions related to services is small. 

<b>Cluster 1:</b> includes many clients. The accumulated gross amount is few but they spent them in paying for services and not so much about office equipment.

<b>Cluster 2:</b> just two data points. They spend a lot of money in card transactions. Their transactions related to services and office equipments are average related to other clients.

<b>Cluster 3:</b> another densely populated segment. They spend little money in card transactions. Their transactions related to office equipments are less than average and their services consumption, by using cards, is just average.

<b>Cluster 4:</b> just one data point that does not spend much momeny but its transactions related to ofice equipment and services are extremely high.

<b>Cluster 5:</b> two clients. They spend a considerable amount of money in card transactions but not related to office equipments or services.

<b>Cluster 6:</b> 44 clients which are ones that spend less money in card transactions, less than average in office equipments and the fewest in services.

<b>Cluster 7:</b> clients which do not spend much money in total per year not in services but more in services.

```python
cost = []
K = range(1,5)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
    kmode.fit_predict(df2014)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```
